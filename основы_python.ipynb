#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
# Your code goes here
def fun(text: str):
    vowels = ["a", "e", "i", "o", "u"]
    answer = 0
    for char in text:
        if char in vowels:
            answer += 1
    return answer

assert fun("ddddddddddaaa") == 3
assert fun("aaauuidddddddd")==6
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n: int) -> int:
    steps = 0
    current = n
    while current >= 10:  # пока число не однозначное
        product = 1
        for digit_char in str(current):
            product *= int(digit_char)
        current = product
        steps =steps+ 1
    return steps

assert magic(39) == 3   
assert magic(4) == 0    
assert magic(999) == 4  
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны иметь одинаковую длину.")
    if len(pred) == 0:
        raise ValueError("Векторы не должны быть пустыми.")
    
    squared_errors = [(p - t) ** 2 for p, t in zip(pred, true)]
    return sum(squared_errors) / len(squared_errors)

assert mse([1, 2, 3], [1, 2, 3]) == 0.0 
assert mse([3, -0.5, 2, 7], [2.5, 0.0, 2, 8]) == 0.375
assert mse([1, 2], [3, 4]) == 4.0

try:
    mse([1, 2], [1])
    assert False, "Должна быть ошибка: разная длина"
except ValueError:
    pass
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factorization(n: int) -> str:
    if n <= 0:
        raise ValueError("Число должно быть положительным.")
    if n == 1:
        return ""  # 1 не имеет простых множителей

    factors = {}
    divisor = 2

    while divisor * divisor <= n:
        while n % divisor == 0:
            factors[divisor] = factors.get(divisor, 0) + 1
            n //= divisor
        divisor += 1

    if n > 1:
        factors[n] = factors.get(n, 0) + 1

    parts = []
    for prime in sorted(factors.keys()):
        exp = factors[prime]
        if exp == 1:
            parts.append(f"({prime})")
        else:
            parts.append(f"({prime}**{exp})")

    return "".join(parts)

def test_prime_factorization():
    assert prime_factorization(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factorization(2) == "(2)"
    assert prime_factorization(17) == "(17)"
    assert prime_factorization(101) == "(101)"
    assert prime_factorization(8) == "(2**3)"      
    assert prime_factorization(81) == "(3**4)"     
    assert prime_factorization(6) == "(2)(3)"      
    assert prime_factorization(30) == "(2)(3)(5)" 
    assert prime_factorization(12) == "(2**2)(3)"  
    assert prime_factorization(100) == "(2**2)(5**2)"
    assert prime_factorization(1) == ""
    assert prime_factorization(2 * 3 * 5 * 7 * 11 * 13) == "(2)(3)(5)(7)(11)(13)"
    try:
        prime_factorization(0)
        assert False, "Должна быть ошибка для n <= 0"
    except ValueError:
        pass

    try:
        prime_factorization(-5)
        assert False, "Должна быть ошибка для отрицательного числа"
    except ValueError:
        pass

if __name__ == "__main__":
    test_prime_factorization()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    # Your code goes here
    if number <= 0:
        return "It is impossible"

    k = 1
    total = 0

    while total < number:
        total = k * (k + 1) * (2 * k + 1) // 6
        if total == number:
            return k
        k += 1

    return "It is impossible"


def test_pyramid():
    
    assert pyramid(1) == 1          
    assert pyramid(5) == 2          
    assert pyramid(14) == 3         
    assert pyramid(30) == 4        
    assert pyramid(55) == 5         
    assert pyramid(91) == 6         
    assert pyramid(140) == 7        

    
    assert pyramid(2) == "It is impossible"
    assert pyramid(10) == "It is impossible"
    assert pyramid(100) == "It is impossible"
    assert pyramid(0) == "It is impossible"
    assert pyramid(-5) == "It is impossible"
    k = 100
    n = k * (k + 1) * (2 * k + 1) // 6
    assert pyramid(n) == k

if __name__ == "__main__":
    test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n: int) -> bool:
    if n <= 0:
        raise ValueError("Число должно быть положительным.")
    
    digits = [int(d) for d in str(n)]
    L = len(digits)
    if L == 1:
        return True
    
    mid = L // 2
    
    if L % 2 == 1:
        # Нечётное: пропускаем одну среднюю цифру (индекс mid)
        left_sum = sum(digits[:mid])
        right_sum = sum(digits[mid + 1:])
    else:
        # Чётное: пропускаем две средние (индексы mid-1 и mid)
        left_sum = sum(digits[:mid - 1])
        right_sum = sum(digits[mid + 1:])
    
    return left_sum == right_sum



def test_is_balanced():
    assert is_balanced(1) == True
    assert is_balanced(9) == True

    assert is_balanced(121) == True      # 1 == 1
    assert is_balanced(12321) == True    # 1+2 == 2+1
    assert is_balanced(12345) == False   # 1+2 != 4+5 (3+6)
    assert is_balanced(13031) == True    # 1+3 == 3+1

    assert is_balanced(1221) == True     # [1] vs [1]
    assert is_balanced(123321) == True   # [1,2] vs [2,1]
    assert is_balanced(123456) == False  # [1,2] vs [5,6] → 3 != 11
    assert is_balanced(11) == True       # [] vs [] → 0 == 0
    assert is_balanced(10) == True       # [] vs [] → 0 == 0 (средние: '1','0')
    assert is_balanced(1234) == False    # [1] vs [4] → 1 != 4
    assert is_balanced(1001) == True     # [1] vs [1]
    assert is_balanced(1010) == False    # [1] vs [0] → 1 != 0

    try:
        is_balanced(0)
        assert False, "Должна быть ошибка"
    except ValueError:
        pass

    try:
        is_balanced(-5)
        assert False, "Должна быть ошибка"
    except ValueError:
        pass



if __name__ == "__main__":
    test_is_balanced()
```
